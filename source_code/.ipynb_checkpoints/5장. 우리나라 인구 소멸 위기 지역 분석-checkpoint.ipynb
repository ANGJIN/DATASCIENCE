## 인구 소멸 위기 지역 계산하고 데이터 정리하기

```python
population['20-39세']= population['20 - 24세'] + population['25 - 29세'] + population['30 - 34세'] + population['35 - 39세']

population['65세이상']= population['65 - 69세'] + population['70 - 74세'] + population['75 - 79세'] + population['80 - 84세'] +\
                        population['85 - 89세'] + population['90 - 94세'] + population['95 - 99세'] + population['100+']
population.head()
```

```python
pop = pd.pivot_table(population, index=['광역시도', '시도'],
                    columns=['구분'],
                    values=['인구수', '20-39세', '65세이상'])
pop
```

```python
pop['소멸비율']= pop['20-39세','여자']/ (pop['65세이상','합계']/2)
pop.head()
```

```python
pop['소멸위기지역']=pop['소멸비율']<1.0
pop.head()
```

```python
si_name= [None]* len(pop)

tmp_gu_dict= {'수원':['장안구', '권선구', '팔달구', '영통구'],
              '성남':['수정구', '중원구', '분당구'],
              '안양':['만안구', '동안구'],
              '안산':['상록구', '단원구'],
              '고양':['덕양구', '일산동구', '일산서구'],
              '용인':['처인구', '기흥구', '수지구'],
              '청주':['상당구', '서원구', '흥덕구', '청원구'],
              '천안':['동남구', '서북구'],
              '전주':['완산구', '덕진구'],
              '포항':['남구', '북구'],
              '창원':['의창구', '성산구', '진해구', '마산합포구', '마산회원구'],
              '부천':['오정구', '원미구', '소사구']}
```

광역시가 아니면서 구를 가지고 있는 시와 그 행정구들

```python
for n in pop.index:
    if pop['광역시도'][n][-3:] not in ['광역시', '특별시', '자치시']: #일반 시 혹은 군일 때 (~~~도)
        #고성 예외처리
        if pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='강원도':
            si_name[n]= '고성(강원)' #강원도 고성
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='경상남도':
            si_name[n]= '고성(경남)' #경상남도 고성
        else:
            si_name[n]= pop['시도'][n][:-1] #일반 시 혹은 군
            
        for keys, values in tmp_gu_dict.items():
            if pop['시도'][n] in values: #일반 시인데 구를 가지는 경우
                if len(pop['시도'][n])==2 : #시도가 두 글자인 경우 전부
                    si_name[n]= keys + ' ' + pop['시도'][n]
                elif pop['시도'][n] in ['마산합포구','마산회원구']:
                    si_name[n]= keys + ' ' + pop['시도'][n][2:-1]
                else:
                    si_name[n]= keys + ' ' + pop['시도'][n][:-1] #이상인경우 구 앞의 글자만 가져옴
    elif pop['광역시도'][n]=='세종특별자치시':
        si_name[n]='세종'
    else: 
        if len(pop['시도'][n])==2:
            si_name[n]= pop['광역시도'][n][:2] + ' ' + pop['시도'][n]
        else:
            si_name[n]= pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
```

각 지역의 ID와 좌표도 모두 확보 완료

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

## 인구 현황에서 여성 인구 비율 확인하기

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = 20.

    datalabel = targetData

    tmp_max = max([ np.abs(min(blockedMap[targetData])), 
                                  np.abs(max(blockedMap[targetData]))])
    vmin, vmax = -tmp_max, tmp_max

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if np.abs(row[targetData]) > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

여성 인구 비율을 확인하기 위해 drawKorea함수 설정을 바꿈. 데이터에 음의 값이 있는지의 여부에 따라 일부 설정을 바꿀 필요가 있기 때문

```python
pop['여성비']= (pop['인구수여자']/pop['인구수합계'] - 0.5)*100
drawKorea('여성비',pop, 'RdBu')
```

여성인구수와 합계를 나눈 후 0.5를 빼서 0이면 50%이도록 계산함. 색상 팔레트가 0을 기준으로 좌우 다른 색상을 갖도록 하기 위함

```python
pop['2030여성비']= (pop['20-39세여자']/pop['20-39세합계'] - 0.5)*100
drawKorea('2030여성비', pop, 'RdBu')
```

20-30대 여성의 20-30대 전체 인구에 대한 비율을 보니 전국적으로 남성 인구 비율이 높은 것을 알수 있다.(전체적으로 지도가 붉다..)
